## Importation Des Librairie

L’objectif de cette partie est d’importer les bibliothèques nécessaires au prétraitement des données, ainsi que de créer des fonctions réutilisables que nous pourrons utiliser à plusieurs reprises.

nous allons aussi créer les deux df pour X_train et X_test sans selection.

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest, f_classif

df_X_train = pd.read_csv('../data/pre_processed/X_train_preprocessed.csv')
df_X_test = pd.read_csv('../data/pre_processed/X_test_preprocessed.csv')

df_y_train = pd.read_csv('../data/pre_processed/y_train.csv')
df_y_test = pd.read_csv('../data/pre_processed/y_test.csv')
```

## Selection Personnel

nous allons commencer par faire une selection personnel des variables que l'on juge les plus pertinnente

```python
selected_columns = [
    "Sex",
    "age",
    "fare",
    "embarked_C", "embarked_Q", "embarked_S",
    "deck_A", "deck_B", "deck_C", "deck_D",
    "deck_E", "deck_F", "deck_G", "deck_nan"
]

# Appliquer sélection
df_X_train_P = df_X_train[selected_columns]
df_X_test_P = df_X_test[selected_columns]
```

## VarianceThreshold

supprime les variables les moins utiles 

```python
selector = VarianceThreshold(threshold=0.05)

# FIT sur TRAIN uniquement
select_X_train_VT = selector.fit_transform(df_X_train)
train_selected_columns = df_X_train.columns[selector.get_support()]

# TRANSFORM sur TEST
select_X_test_VT = selector.transform(df_X_test)
test_selected_columns = train_selected_columns

# Recréer DataFrame
df_X_train_VT = pd.DataFrame(select_X_train_VT, columns=train_selected_columns)
df_X_test_VT = pd.DataFrame(select_X_test_VT, columns=test_selected_columns)
```

## SelectKBest

garde les variables les k meilleures variables

```python
selector = SelectKBest(score_func=f_classif, k=7)

# FIT sur TRAIN
select_X_train_KB = selector.fit_transform(df_X_train, df_y_train.values.ravel())
train_selected_columns = df_X_train.columns[selector.get_support()]

# TRANSFORM sur TEST
select_X_test_KB = selector.transform(df_X_test)
test_selected_columns = train_selected_columns

# Recréer DataFrame
df_X_train_KB = pd.DataFrame(select_X_train_KB, columns=train_selected_columns)
df_X_test_KB = pd.DataFrame(select_X_test_KB, columns=test_selected_columns)
```

## exportation des df

```python
export_path = "../data/selected/"

df_X_train.to_csv(f"{export_path}X_train_N.csv", index=False)
df_X_test.to_csv(f"{export_path}X_test_N.csv", index=False)

df_X_train_P.to_csv(f"{export_path}X_train_P.csv", index=False)
df_X_test_P.to_csv(f"{export_path}X_test_P.csv", index=False)

df_X_train_VT.to_csv(f"{export_path}X_train_VT.csv", index=False)
df_X_test_VT.to_csv(f"{export_path}X_test_VT.csv", index=False)

df_X_train_KB.to_csv(f"{export_path}X_train_KB.csv", index=False)
df_X_test_KB.to_csv(f"{export_path}X_test_KB.csv", index=False)
```
